```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
```

```python
x = np.linspace(0, 2 * np.pi, 1000)
y = np.sin(x)
```

```python
fig, ax = plt.subplots()
line, = ax.plot(x, y)
ax.set_ylim(-1.5, 1.5)
```

```python
def update(frame):
    line.set_ydata(np.sin(x + frame / 10))
    return line,
```

```python
ani = FuncAnimation(fig, update, frames=100, interval=50, blit=True)
ani
```
