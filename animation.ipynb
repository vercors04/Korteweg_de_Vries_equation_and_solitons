```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
delay=20 # in milliseconds between frames
nframe = 100 # number of frames for “one round of animation”
```

```python
fig, ax = plt.subplots()
line, = ax.plot([])
xmin=0
xmax=2*np.pi
nx=100
x=np.linspace(xmin,xmax,nx)
w=2*np.pi
ax.set_xlim(xmin,xmax)
ax.set_ylim(-1.1, 1.1)
def animate(frame_num):
    y = np.sin(x + w * frame_num/nframe)
    line.set_data((x, y))
    return line
anim = FuncAnimation(fig, animate, frames=nframe, interval=delay)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython import display
plt.rcParams['animation.ffmpeg_path'] = '/usr/bin/ffmpeg'
delay=20 # in milliseconds between frames
nframe = 100 # number of frames for “one round of animation”
```

```python
fig, ax = plt.subplots()
line, = ax.plot([])
xmin=0
xmax=2*np.pi
nx=100
x=np.linspace(xmin,xmax,nx)
w=2*np.pi
ax.set_xlim(xmin,xmax)
ax.set_ylim(-1.1, 1.1)
def animate(frame_num):
    y = np.sin(x + w * frame_num/nframe)
    line.set_data((x, y))
    return line
anim = FuncAnimation(fig, animate, frames=nframe, interval=delay)

video = anim.to_html5_video()
html = display.HTML(video)
display.display(html)
plt.close()
```
