```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
c = 0 # vitesse de propagation du soliton
eps = 0.2
mu = 0.1
```

```python
# Fonction mathematique sech
def sech(x):
    return 1 / np.cosh(x)
```

```python
# x : position 
# t : temps
def sol_an(x,t):
    xi = x-c*t
    xi_0=0
    return 3*c/eps * sech(1/2 * math.sqrt(c/mu) * (xi-xi_0))**2
```

```python
#graphique
x = np.linspace(-20, 20, 1000)
u_t0 = sol_an(x, 0)
fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(x, u_t0, label='Soliton t=0', color='blue', linewidth=2)
ax.set_title("Condition Initiale du Soliton (Profil KdV)")
plt.show()
```
