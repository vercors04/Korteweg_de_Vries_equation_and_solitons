```python
import numpy as np
```

```python
c = 0 # vitesse de propagation du soliton
eps = 0.2
mu = 0.1
```

```python
# Fonction mathematique sech
def sech(x):
    return 1 / np.cosh(x)
```

```python
# x : position 
# t : temps
def sol_an(x,t):
    xi = x-c*t
    xi_0=0
    return 3*c/eps * sech(1/2 * sqrt(c/mu) * (xi-xi_0))**2
```
